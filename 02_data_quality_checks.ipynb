```python
df.isnull().sum()
```


- created features have significant # of null values 
- primary suspected cause: due to the cumulative nature of metrics (i.e. last 9 matches will be null for first 9 records)
- secondary suspected cause: team 1 vs team 2 comparison metrics (if one team has a null value, both will)

To test these hypotheses, we can exclude first 9 outcomes of each recognised country.

```python
for i in df["away_team"].unique():
    df_temp = df[(df.away_team==i)|(df.home_team==i)]
```
